```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("companion_plant_wikipedia - my_version.tsv", sep="\t")
df.head()
```

```python
for col in ["Common name", "Scientific name"]:
    df[col] = df[col].str.replace("\*", "")
```

```python
df.head()
```

```python
for col in ["Helps", "Helped by", "Attracts", "-Repels/+distracts", "Avoid"]:
    df[col] = df[col].str.replace("\[.+?\]", "")
```

```python
df.head()
```

```python
for col in ["Common name", "Helps", "Helped by", "Attracts", "-Repels/+distracts", "Avoid", "Comments"]:
    df[col] = df[col].str.lower()

df.head()
```

```python
df.to_csv("companion_plant_wikipedia_after_prepro.tsv", sep="\t", index=False)
```
